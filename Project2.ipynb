##**Keyword VS Sentiments**

####**Day** 

```python
import matplotlib.pyplot as plt
import seaborn as sns
def percentage(x):
  return [(x[0]*100)/(sum(x)) , (x[1]*100)/(sum(x))]
colors = ['#F4F7F0', '#BFD2D0']
names = ['Positive', 'Negative']

x1 = [2029, 620]
x2 = [1071, 328]
x3 =[392, 155]
plt.figure(figsize=(18, 4))
plt.subplot(131)
plt.bar(names, x1, color=('#A4E5E0','#37BEB0'))
plt.ylabel('Number of Tweets')
plt.xlabel('Sentiments')
plt.subplot(132)
plt.bar(names, x2, color=('#A4E5E0','#37BEB0'))
plt.ylabel('Number of Tweets')
plt.xlabel('Sentiments')
plt.subplot(133)
plt.bar(names, x3, color=('#A4E5E0','#37BEB0'))
plt.ylabel('Number of Tweets')
plt.xlabel('Sentiments')
plt.suptitle('Day')
plt.show()

plt.figure(figsize=(20,4))
plt.subplot(131)
plt.pie(percentage(x1), labels=['Positive', 'Negative'], colors=colors, autopct='%1.1f%%')
plt.subplot(132)
plt.pie(x2, labels=['Positive', 'Negative'], colors=colors, autopct='%1.1f%%')
plt.subplot(133)
plt.pie(x3, labels=['Positive', 'Negative'], colors=colors, autopct='%1.1f%%')
plt.suptitle('Setiment Percentage')
plt.show()
```
